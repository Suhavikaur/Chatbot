```python
from tkinter import *

# GUI
root = Tk()
root.title("Chatbot")

BG_GRAY = "#E1AFD1"
BG_COLOR = "#FFE6E6"
TEXT_COLOR = "#070F2B"

FONT = "Helvetica 14"
FONT_BOLD = "Helvetica 13 bold"

# Send function
def send():
	send = "You : " + e.get()
	txt.insert(END, "\n" + send)

	user = e.get().lower()
#Chatbot Response Training
	if (user == "hello"):
		txt.insert(END, "\n" + "Bot : Hi there, how can I help?")

	elif (user == "hi" or user == "hii" or user == "hiiii" or user == "hi there"):
		txt.insert(END, "\n" + "Bot : Hi there, what can I do for you?")
	elif (user == "suggest flight booking sites" or user == "suggest some good flight booking sites" or user == "flight booking sites names"):
		txt.insert(END, "\n" + "Bot : MakemyTrip, Ixigo, Agoda, EasemyTrip")
	elif (user == "suggest me youtube channels to learn ai" or user == "suggest youtube channels to learn ai" or user == "best ai tutorials on youtube"):
		txt.insert(END, "\n" + "Bot : DeepLearning AI, Machine Learning Street Talk")

	elif (user == "how are you" or user == "how do you do?"):
		txt.insert(END, "\n" + "Bot : fine! and you")
	elif (user == "what's your favourite cartoon" or user == "your favourite cartoon?"):
		txt.insert(END, "\n" + "Bot : Tom and Jerry!")

	elif (user == "fine" or user == "i am good" or user == "i am doing good" or user == "i am fine"):
		txt.insert(END, "\n" + "Bot : Great! how can I help you.")

	elif (user == "thanks" or user == "thank you" or user == "now its my time" or user == "bye"):
		txt.insert(END, "\n" + "Bot : My pleasure !")

	elif (user == "what do you sell" or user == "what kinds of items are there" or user == "have you something"):
		txt.insert(END, "\n" + "Bot : We have coffee and tea")

	elif (user == "tell me a joke" or user == "tell me something funny" or user == "crack a funny line"):
		txt.insert(
			END, "\n" + "Bot : What did the buffalo say when his son left for college? Bison.! " or "Bot : Hahaha")

	elif (user == "goodbye" or user == "see you later" or user == "see yaa"):
		txt.insert(END, "\n" + "Bot : Have a nice day!")
	elif (user == "suggest me some good movies" or user == "suggest movies" or user == "suggest some good movies"):
		txt.insert(END, "\n" + "Bot : Kill Bill, Pulp Fiction, The Harry Potter Series and MCU Movies are really good!")
	elif (user == "suggest me some solo travel destinations" or user == "suggest solo travel destinations" or user == "suggest solo travel places"):
		txt.insert(END, "\n" + "Bot : Himachal Pradesh, Kerala, Goa, Kashmir and so on are some good places for solo travel" + "\n" + "https://www.holidify.com/collections/solo-travel-india")
    
	else:
		txt.insert(END, "\n" + "Bot : Sorry! I didn't understand that")

	e.delete(0, END)


lable1 = Label(root, bg=BG_COLOR, fg=TEXT_COLOR, text="Welcome", font=FONT_BOLD, pady=10, width=10, height=1).grid(
	row=0)

txt = Text(root, bg=BG_COLOR, fg=TEXT_COLOR, font=FONT, width=100)
txt.grid(row=1, column=0, columnspan=2)

scrollbar = Scrollbar(txt)
scrollbar.place(relheight=1, relx=0.974)

e = Entry(root, bg="#AD88C6", fg=TEXT_COLOR, font=FONT, width=55)
e.grid(row=2, column=0)

send = Button(root, text="Send", font=FONT_BOLD, bg=BG_GRAY,
			command=send).grid(row=2, column=1)

root.mainloop()
```
